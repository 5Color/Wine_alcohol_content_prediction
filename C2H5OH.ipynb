```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from lightgbm import LGBMRegressor
```

```python

data = pd.read_csv("WineQT.csv")
```

```python
data
```

```python

x = data[['fixed acidity','volatile acidity','citric acid',
          'residual sugar','chlorides','free sulfur dioxide',
          'total sulfur dioxide','density','pH','sulphates','quality']]
y = data['alcohol']
```

```python
y.shape
```

```python
train_input, test_input, train_target, test_target = train_test_split(x, y, test_size=0.2)
train_input.shape
```

```python
model = LGBMRegressor()
model.fit(train_input, train_target)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, root_mean_squared_error
pred = model.predict(test_input)
print("MAE : ", mean_absolute_error(pred, test_target))
print("MSE : ", mean_squared_error(pred, test_target))
print("R^2 ê²°ì •ê³„ìˆ˜ : ", r2_score(pred, test_target))
print("RMSE : ", root_mean_squared_error(pred, test_target))
```

```python
reg = LinearRegression()
reg.fit(train_input, train_target)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, root_mean_squared_error
pred = reg.predict(test_input)
print("MAE : ", mean_absolute_error(pred, test_target))
print("MSE : ", mean_squared_error(pred, test_target))
print("R^2 ê²°ì •ê³„ìˆ˜ : ", r2_score(pred, test_target))
print("RMSE : ", root_mean_squared_error(pred, test_target))
```

```python
from sklearn.ensemble import RandomForestRegressor
randomforest = RandomForestRegressor()

randomforest.fit(train_input, train_target)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, root_mean_squared_error
pred = randomforest.predict(test_input)
print("MAE : ", mean_absolute_error(pred, test_target))
print("MSE : ", mean_squared_error(pred, test_target))
print("R^2 ê²°ì •ê³„ìˆ˜ : ", r2_score(pred, test_target))
print("RMSE : ", root_mean_squared_error(pred, test_target))
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(train_input, train_target)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, root_mean_squared_error
pred = xgb.predict(test_input)
print("MAE : ", mean_absolute_error(pred, test_target))
print("MSE : ", mean_squared_error(pred, test_target))
print("R^2 ê²°ì •ê³„ìˆ˜ : ", r2_score(pred, test_target))
print("RMSE : ", root_mean_squared_error(pred, test_target))
```

```python
from sklearn.ensemble import StackingRegressor
base_learners = [
    ('lr', LinearRegression()),              # ì„ í˜• baseline
    ('rf', RandomForestRegressor(n_estimators=200, random_state=42)),  # Bagging ê³„ì—´
    ('lgbm', LGBMRegressor(n_estimators=300, random_state=42))         # Boosting ê³„ì—´
]
stacking = StackingRegressor(estimators=base_learners, final_estimator=LinearRegression())

stacking.fit(train_input, train_target)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, root_mean_squared_error
pred = stacking.predict(test_input)
print("MAE : ", mean_absolute_error(pred, test_target))
print("MSE : ", mean_squared_error(pred, test_target))
print("R^2 ê²°ì •ê³„ìˆ˜ : ", r2_score(pred, test_target))
print("RMSE : ", root_mean_squared_error(pred, test_target))
```
